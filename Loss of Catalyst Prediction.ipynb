```python
import pandas as pd
from pandas import json_normalize  # Import json_normalize function from pandas.io.json
import json

# Load the JSON data
with open('spent-cat-library.v1.json', 'r') as file:
    data = json.load(file)

# Normalize each array of objects
df_support = json_normalize(data['spent_cat_support'])
df_metal = json_normalize(data['spent_cat_metal'])
df_hazard = json_normalize(data['spent_cat_hazard'])
df_bulk_density = json_normalize(data['spent_cat_bulk_density'])

# Merge all DataFrames into one
df = pd.concat([df_support, df_metal, df_hazard, df_bulk_density], ignore_index=True)

# Display the DataFrame
print(df.head())
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train_scaled,y_train)
```

```python
y_pred = lr.predict(x_test_scaled)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_pred,y_test)
```

```python
regressor = RandomForestRegressor(n_estimators=10, random_state=0)
```

```python
regressor.fit(x_train_scaled, y_train)
```

```python
re = regressor.predict(x_test_scaled)
```

```python
r2 = r2_score(y_test,re)
```
